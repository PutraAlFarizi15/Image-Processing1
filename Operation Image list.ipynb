```python
#Negative
import cv2
def negative(img):
    row = img.shape[0]
    column = img.shape[1]
    bright = img.copy()
    for i in range(row):
        for j in range(column):
            bright[i, j] = 255 - img[i, j]
    return bright

img = cv2.imread('Rose.jpg')
img_grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
citra_n = negative(img_grayscale)
cv2.imshow('Grayscale', img_grayscale)
cv2.imshow('Negative', citra_n)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Brightness
import cv2
def clipping(intensitas):
    if intensitas < 0:
        return 0
    if intensitas > 255:
        return 255
    return intensitas

def brightness_image_grayscale(img, kecerahan):
    row = img.shape[0]
    column = img.shape[1]
    bright = img.copy()
    for i in range(row):
        for j in range(column):
            bright[i, j] = clipping(img[i, j] + kecerahan)
    return bright

img = cv2.imread('Rose.jpg')
img_grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_brightness_90 = brightness_image_grayscale(img_grayscale, 90)
img_brightness_n90 = brightness_image_grayscale(img_grayscale, -90)
cv2.imshow('Grayscale', img_grayscale)
cv2.imshow('Brightness 90', img_brightness_90)
cv2.imshow('Brightness -90', img_brightness_n90)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#sharp
import cv2
def convolution(image, mask):
    row = image.shape[0]
    column = image.shape[1]
    image_result = image.copy()
    for i in range(1, row-2):
        for j in range(1, column-2):
            image_result[i][j]=(
                image[i-1][j-1]*mask[0][0] +
                image[i-1][j]*mask[0][1] +
                image[i-1][j+1]*mask[0][2] +
                image[i][j-1]*mask[1][0] +
                image[i][j]*mask[1][1] +
                image[i][j+1]*mask[1][2] +
                image[i+1][j-1]*mask[2][0] +
                image[i+1][j]*mask[2][1] +
                image[i+1][j+1]*mask[2][2])
    return image_result

def sharp(image):
    mask = [[0, -1, 0],
             [-1, 5, -1],
             [0, -1, 0]]
    return convolution(image, mask)

img = cv2.imread('Rose.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
sharp = sharp(img_gray)
cv2.imshow('Grayscale', img_gray)
cv2.imshow('Sharp', sharp)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#smooth
import cv2
def convolution(image, mask):
    row = image.shape[0]
    column = image.shape[1]
    image_result = image.copy()
    for i in range(1, row-2):
        for j in range(1, column-2):
            image_result[i][j]=(
                image[i-1][j-1]*mask[0][0] +
                image[i-1][j]*mask[0][1] +
                image[i-1][j+1]*mask[0][2] +
                image[i][j-1]*mask[1][0] +
                image[i][j]*mask[1][1] +
                image[i][j+1]*mask[1][2] +
                image[i+1][j-1]*mask[2][0] +
                image[i+1][j]*mask[2][1] +
                image[i+1][j+1]*mask[2][2])
    return image_result

def smooth(image):
    mask = [[1/9, 1/9, 1/9],
            [1/9, 1/9, 1/9],
            [1/9, 1/9, 1/9]]
    return convolution(image, mask)

img = cv2.imread('Rose.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
smooth = smooth(img_gray)
cv2.imshow('Grayscale', img_gray)
cv2.imshow('Smooth', smooth)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Gradient
import cv2
def gradient(image):
    gx = [-1, 1]
    gy = [1, -1]
    row = image.shape[0]
    column = image.shape[1]
    image_result = image.copy()
    for i in range(0, row-2):
        for j in range(0, column-2):
            image_result_x = image[i][j]*gx[0] + image[i][j+1]*gx[1]
            image_result_y = image[i][j]*gy[0] + image[i+1][j]*gy[1]
            image_result[i][j] = abs(image_result_x)+abs(image_result_y)
    return image_result

img = cv2.imread('Rose.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gradient = gradient(img_gray)
cv2.imshow('Grayscale', img_gray)
cv2.imshow('Gradien', gradient)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Center Difference
import cv2
def center_difference(image):
    gx = [-1, 0, 1]
    gy = [1, 0, -1]
    row = image.shape[0]
    column = image.shape[1]
    image_result = image.copy()
    for i in range(0, row-3):
        for j in range(0, column-3):
            image_result_x = image[i][j]*gx[0] + image[i][j+1]*gx[1] + image[i][j+2]*gx[2]
            image_result_y = image[i][j]*gy[0] + image[i+1][j]*gy[1] + image[i+2][j]*gy[2]
            image_result[i][j] = abs(image_result_x)+abs(image_result_y)
    return image_result

img = cv2.imread('Rose.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cd = center_difference(img_gray)
cv2.imshow('Grayscale', img_gray)
cv2.imshow('Center Difference', cd)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Sobel
import cv2
def sobel(image):
    sx = [[-1, 0, 1],
         [-2, 0, 2],
         [-1, 0, 1]]
    
    sy = [[1, 2, 1],
          [0, 0, 0],
          [-1, -2, -1]]
    
    row = image.shape[0]
    column = image.shape[1]
    image_result = image.copy()
    for i in range(1, row-2):
        for j in range(1, column-2):
            image_result_x=(
                image[i-1][j-1]*sx[0][0] +
                image[i][j-1]*sx[1][0] +
                image[i+1][j-1]*sx[2][0] +
                image[i-1][j]*sx[0][1] +
                image[i][j]*sx[1][1] +
                image[i+1][j]*sx[2][1] +
                image[i-1][j+1]*sx[0][2] +
                image[i][j+1]*sx[1][2] +
                image[i+1][j+1]*sx[2][2])
            
            image_result_y=(
                image[i-1][j-1]*sy[0][0] +
                image[i-1][j]*sy[0][1] +
                image[i-1][j+1]*sy[0][2] +
                image[i][j-1]*sy[1][0] +
                image[i][j]*sy[1][1] +
                image[i][j+1]*sy[1][2] +
                image[i+1][j-1]*sy[2][0] +
                image[i+1][j]*sy[2][1] +
                image[i+1][j+1]*sy[2][2])
           
            image_result[i][j] = abs(image_result_x) + abs(image_result_y)
    return image_result

img = cv2.imread('Rose.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
sobel = sobel(img_gray)
cv2.imshow('Grayscale', img_gray)
cv2.imshow('Sobel', sobel)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Roberts
import cv2
def roberts(image):
    sx = [[1, 0],
          [0, -1]]
    
    sy = [[0, 1],
          [-1, 0]]
    
    row = image.shape[0]
    column = image.shape[1]
    image_result = image.copy()
    for i in range(0, row-2):
        for j in range(0, column-2):
            image_result_x=(
                image[i][j]*sx[0][0] +
                image[i+1][j]*sx[1][0] +
                image[i][j+1]*sx[0][1] +
                image[i+1][j+1]*sx[1][1])
            
            image_result_y=(
                image[i][j]*sy[0][0] +
                image[i][j+1]*sy[0][1] +
                image[i+1][j]*sy[1][0] +
                image[i+1][j+1]*sy[1][1])
            image_result[i][j] = abs(image_result_x) + abs(image_result_y)
    return image_result

img = cv2.imread('Rose.jpg')  
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
roberts = roberts(img_gray)
cv2.imshow('Grayscale', img_gray)
cv2.imshow("Roberts", roberts)
cv2.waitKey(0)
cv2.destroyAllWindows()
                        
```

```python
#Laplacian
import cv2
def convolution(image, mask):
    image_result = image.copy()
    row = image.shape[0]
    column = image.shape[1]
    image_result = image.copy()
    for i in range(1, row-2):
        for j in range(1, column-2):
            image_result[i][j]=(
                image[i-1][j-1]*mask[0][0] +
                image[i-1][j]*mask[0][1] +
                image[i-1][j+1]*mask[0][2] +
                image[i][j-1]*mask[1][0] +
                image[i][j]*mask[1][1] +
                image[i][j+1]*mask[1][2] +
                image[i+1][j-1]*mask[2][0] +
                image[i+1][j]*mask[2][1] +
                image[i+1][j+1]*mask[2][2])
    return image_result

def gaussian(image):
    gauss = [[1/16, 2/16, 1/16],  
             [2/16, 4/16, 2/16],  
             [1/16, 2/16, 1/16]]
    return convolution(image, gauss)

def laplacian(img, index):
    image = gaussian(img)
    if(index==0):
        lp_kernel = [[0, 1, 0],
                   [1, -4, 1],
                   [0, 1, 0]]
        return convolution(image, lp_kernel)
    elif(index==1):
        lp_kernel = [[-1, -1, -1],
                     [-1, 8, -1],
                     [-1, -1, -1]]
        return convolution(image, lp_kernel)
    elif(index==2):
        lp_kernel = [[1, -2, 1],
                     [-2, 4, -2],
                     [1, -2, 1]]
        return convolution(image, lp_kernel)
    elif(index==3):
        lp_kernel = [[1, 4, 1],
                     [4, -20, 4],
                     [1, 4, 1]]
        return convolution(image, lp_kernel)
    return img

img = cv2.imread('Rose.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
laplacian_0 = laplacian(img_gray, 0)
laplacian_1 = laplacian(img_gray, 1)
laplacian_2 = laplacian(img_gray, 2)
laplacian_3 = laplacian(img_gray, 3)
cv2.imshow("Grayscale", img_gray)
cv2.imshow('Laplacian 0', laplacian_0)
cv2.imshow('Laplacian 1', laplacian_1)
cv2.imshow('Laplacian 2', laplacian_2)
cv2.imshow('Laplacian 3', laplacian_3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Compass
import cv2
def convolution(image, mask):
    image_result = image.copy()
    row = image.shape[0]
    column = image.shape[1]
    image_result = image.copy()
    for i in range(1, row-2):
        for j in range(1, column-2):
            image_result[i][j]=(
                image[i-1][j-1]*mask[0][0] +
                image[i-1][j]*mask[0][1] +
                image[i-1][j+1]*mask[0][2] +
                image[i][j-1]*mask[1][0] +
                image[i][j]*mask[1][1] +
                image[i][j+1]*mask[1][2] +
                image[i+1][j-1]*mask[2][0] +
                image[i+1][j]*mask[2][1] +
                image[i+1][j+1]*mask[2][2])
    return image_result

def compass(image, direction):
    if(direction=='north'):
        north = [[1, 1, 1],
                [1, -2, 1],
                [-1, -1, -1]]
        return convolution(image, north)
    elif(direction=='northeast'):
        northeast = [[1, 1, 1],
                     [-1, -2, 1],
                     [-1, -1, 1]]
        return convolution(image, northeast)
    elif(direction=='east'):
        east = [[-1, 1, 1],
                [-1, -2, 1],
                [-1, 1, 1]]
        return convolution(image, east)
    elif(direction=='southeast'):
        southeast = [[-1, -1, 1],
                     [-1, -2, 1],
                     [1, 1, 1]]
        return convolution(image, southeast)
    elif(direction=='south'):
        south = [[-1, -1, -1],
                [1, -2, 1],
                [1, 1, 1]]
        return convolution(image, south)
    elif(direction=='southwest'):
        southwest = [[1, -1, -1],
                [1, -2, -1],
                [1, 1, 1]]
        return convolution(image, southwest)
    elif(direction=='west'):
        west = [[1, 1, -1],
                [1, -2, -1],
                [1, 1, -1]]
        return convolution(image, west)
    elif(direction=='northwest'):
        northwest = [[1, 1, 1],
                [1, -2, -1],
                [1, -1, -1]]
        return convolution(image, northwest)
    return image

img = cv2.imread('Rose.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
compass_n = compass(img_gray, 'north')
compass_ne = compass(img_gray, 'northeast')
compass_e = compass(img_gray, 'east')
compass_se = compass(img_gray, 'southeast')
compass_s = compass(img_gray, 'south')
compass_sw = compass(img_gray, 'southwest')
compass_w = compass(img_gray, 'west')
compass_nw = compass(img_gray, 'northwest')
cv2.imshow("Grayscale", img_gray)
cv2.imshow("Compass North", compass_n)
cv2.imshow("Compass Northeast", compass_ne)
cv2.imshow("Compass East", compass_e)
cv2.imshow("Compass Southeast", compass_se)
cv2.imshow("Compass South", compass_s)
cv2.imshow("Compass Southwest", compass_sw)
cv2.imshow("Compass West", compass_w)
cv2.imshow("Compass Northwest", compass_nw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##  3. Binerisasi Citra

```python
import cv2
def to_citra_biner(img, t):
    row = img.shape[0]
    column = img.shape[1]
    binary = img.copy()
    for i in range(row):
        for j in range(column):
            if((img[i][j])<t):
                binary[i, j] = 0
            else:
                binary[i, j] = 255
    return binary

img = cv2.imread('Rose.jpg')
img_grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
binary = to_citra_biner(img_grayscale, 120)
cv2.imshow('Grayscale',img_grayscale)
cv2.imshow('Binary',binary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 4. Convert Color RGB To HSI

```python
import cv2
import numpy as np
import math

def rgb_to_hsi(img):

    with np.errstate(divide='ignore', invalid='ignore'):

        #Load image with 32 bit floats as variable type
        bgr = np.float32(img)/255

        #Separate color channels
        blue = bgr[:,:,0]
        green = bgr[:,:,1]
        red = bgr[:,:,2]

        def calc_intensity(red, blue, green):
            return np.divide(blue + green + red, 3)

        def calc_saturation(red, blue, green):
            minimum = np.minimum(np.minimum(red, green), blue)
            saturation = 1 - (3 / (red + green + blue) * minimum)
            return saturation

        def calc_hue(red, blue, green):
            hue = np.copy(red)
            for i in range(0, blue.shape[0]):
                for j in range(0, blue.shape[1]):
                    hue[i][j] = (0.5 * ((red[i][j] - green[i][j]) + (red[i][j] - blue[i][j])) / 
                                math.sqrt((red[i][j] - green[i][j])**2 +
                                        ((red[i][j] - blue[i][j]) * (green[i][j] - blue[i][j]))))
                    hue[i][j] = math.acos(hue[i][j])

                    if blue[i][j] <= green[i][j]:
                        hue[i][j] = hue[i][j]
                    else:
                        hue[i][j] = ((360 * math.pi) / 180.0) - hue[i][j]
            return hue

        #Merge channels into picture and return image
        hsi = cv2.merge((calc_hue(red, blue, green), calc_saturation(red, blue, green), calc_intensity(red, blue, green)))
        return hsi

img = cv2.imread('fruit4.jpg')
hsi = rgb_to_hsi(img)   
cv2.imshow('Image', img)
cv2.imshow('HSI Color', hsi)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
